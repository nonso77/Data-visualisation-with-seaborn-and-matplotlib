# DATA VISUALISATION OF THE FUEL DATASET USING SEABORN AND MATPLOTLIB

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

_Below we read in our csv file from our working directory using the pandas library and alo check for the first 20 columns and rows_

```python
df=pd.read_csv('fuel.csv')
```

**DATA VISUALISATIONS**

_Below we created some visualisations using our data from the tables_

```python
sns.countplot(data=df,x='fuelType',color='red');
plt.title('fuelType per quantity')
```

_the above chart shows that the fuel type is the premium gasoline and the lwest is the natural gas_

```python
##created a bar chart using seaborn
sns.countplot(data=df,x='cylinders',color='red');
plt.title('cylindertype per quantity');
```

_The above chart shows that our highest cylinder per quantity is cylinder number 6 and the lowest is cylinder number 2_

```python
## created a scattered plot using matplotlib library
sns.regplot(data=df,x='co2 emission',y='displacement');
plt.title('co2 emissions per displacement')
plt.xlabel('co2 emissions')
plt.ylabel('displacement');
```

_using the scattered plot above to see the relationship between the co2 emissions and displacement_
